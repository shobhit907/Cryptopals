```python
def base64tobin(s):
    bin1=""
    for x in s:
        bin2=""
        if(x>='A' and x<='Z'):
            bin2=bin(ord(x)-ord('A'))
        elif(x>='a' and x<='z'):
            bin2=bin(ord(x)-ord('a')+26)
        elif(x>='0' and x<='9'):
            bin2=bin(ord(x)-ord('0')+52)
        elif(x=='+'):
            bin2=bin(62)
        elif(x=='/'):
            bin2=bin(63)
        bin2=bin2[2:]
        while(len(bin2)<6):
            bin2='0'+bin2
        bin1+=bin2
    return bin1
def bintoascii(bin1):
    while(len(bin1)%8!=0):
        bin1='0'+bin1
    ret=""
    for i in range(0,len(bin1),8):
        ret+=chr(int(bin1[i:i+8],2))
    return ret
def base64toascii(s):
    bin1=base64tobin(s)
    return bintoascii(bin1)
```

```python
def do_xor(s1,s2):
    s3=""
    for i in range(0,len(s1)):
        if(s1[i]=='0' and s2[i]=='1' or s1[i]=='1' and s2[i]=='0'):
            s3+='1'
        else:
            s3+='0'
    return s3
```
